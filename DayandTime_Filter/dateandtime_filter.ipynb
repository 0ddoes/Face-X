# **Converting the Image into greyscale Image**

```python
img = cv2.imread('/content/sample1.png')
def greyscale(img):
    greyscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return greyscale
a1 = greyscale(img)
plt.imshow(cv2.cvtColor(a1, cv2.COLOR_BGR2RGB))
```

# **Procedure**


*   Step 1 : Finding the day and time using dateandtime module
*   Step 2 : adjusting the font size
*   Step 3 : Finding the postion of time to place on the photo
*   Step 4 : Get back the image to OpenCV









```python
import datetime
import math
import time
import pytz 

#Finding the current_time
current_time = datetime.datetime.now(pytz.timezone('Asia/Kolkata')) 
current_time = current_time.strftime("%H:%M")
t = current_time

#Finding the present_day
from datetime import datetime
s = datetime.today().strftime('%A')


from PIL import ImageFont, ImageDraw, Image  
import cv2  
import numpy as np  
   
text_of_day = s 
text_of_time = t  

#Load image in OpenCV  
image = cv2.imread("/content/sample1.png")  
width_image = image.shape[1]
height_image = image.shape[0]

#Convert the image to RGB (OpenCV uses BGR)  
cv2_im_rgb = cv2.cvtColor(a1,cv2.COLOR_BGR2RGB)  
   
#Pass the image to PIL  
pil_im = Image.fromarray(cv2_im_rgb)  
   
draw = ImageDraw.Draw(pil_im)
col = ImageColor.getrgb("skyblue")

#adjusting the font size
fontsize = 1  # starting font size

# portion of image width you want text width to be
img_fraction = 0.50
font = ImageFont.truetype("/content/Bold & Stylish Calligraphy.ttf", fontsize)
while (int)(font.getsize(text_of_day)[0]) < (int)(img_fraction*image.shape[0]):
    # iterate until the text size is just larger than the criteria
    fontsize += 1
    font = ImageFont.truetype("/content/Bold & Stylish Calligraphy.ttf", fontsize)

# optionally de-increment to be sure it is less than criteria
fontsize -= 1
font = ImageFont.truetype("/content/Bold & Stylish Calligraphy.ttf", fontsize)  

height_of_text_for_time = font.getsize(text_of_time)[1]
width_of_text_for_time = font.getsize(text_of_time)[0] 

#calculating_coordinates_of_time
position_of_time = (width_image - width_of_text_for_time , height_image - height_of_text_for_time) 

# Draw the text  
draw.text((0,0), text_of_day, font=font,fill = col)  
draw.text(position_of_time, text_of_time, font=font,fill = col)
# Get back the image to OpenCV  
cv2_im_processed = cv2.cvtColor(np.array(pil_im), cv2.COLOR_RGB2BGR)  
plt.imshow(cv2.cvtColor(cv2_im_processed, cv2.COLOR_BGR2RGB))
```
