```python
from tensorflow import keras
model = keras.models.load_model('Gender_final.h5')
import cv2
detect=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -------------------------------------Real Time version----------------------------------- 
c = cv2.VideoCapture(0)
while(True):
    ret, img = c.read()
    gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = detect.detectMultiScale(gray, 1.3, 4)
    full  = img
    for (x,y,w,h) in faces:
        img  = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        img1 = img[y:y+h, x:x+w]
        face = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
        face = cv2.resize(face, (32,32) )
        face = np.expand_dims(face, axis=0)
        arr  = model.predict(face)
        if arr[0][0] < 0.3 :
            gend = 'Male'
        else:
            gend = 'Female  '
        cv2.putText(full,'Gender = '+gend+str(arr[0][0]), (x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(100, 80, 0, 255),3)
    cv2.imshow('img',full)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
c.release()
cv2.destroyAllWindows()
```

```python
#-------------------------------------USE if the Face is un cropped----------------------------------
img  = cv2.imread('test_data//4c.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = detect.detectMultiScale(gray, 1.3, 5)
full = img
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)
    img1  = img[y:y+h, x:x+w]
    plt.imshow(img1)
    plt.show()
    face = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    face = cv2.resize(face, (32,32) )
    face = np.expand_dims(face, axis=0)
    result = model.predict(face)
    if result[0][0] < 0.5 :
        gender = 'Male'
    else:
        gender = 'Female'
    print(gender)
    #full = cv2.cvtColor(full, cv2.COLOR_RGB2BGR)
    cv2.imshow('img',full)
    cv2.waitKey(0) 
    cv2.destroyAllWindows() 
#Real Age 35
```
